## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO

import numpy as np
import random

n_iteraciones = 1000

def lanzada_de_dado():
    return random.randint(1,6)

contador = 0

for i in range(n_iteraciones):
    lanzamiento_1 = lanzada_de_dado()
    lanzamiento_2 = lanzada_de_dado()
    suma = lanzamiento_1 + lanzamiento_2
    if suma > 7 or suma%2 == 0:
        contador += 1
print("La probabilidad experimental fue de: "+str(contador/n_iteraciones))
print("La probabilidad laplaciana fue de: "+str(24/36))


#S = {2-1,3-2,4-3,5-4,6-5,7-6,8-5,9-4,10-3,11-2,12-1} =36 (Sample Space)
#E = {2-1,4-3,6-6,8-5,10-3,12-1,9-4,11-2} =25 (Evento)
```

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"
print("Caja de bolitas:" ,ball_box)

# TODO

def sacar_una_bolita():
    return ball_box[random.randint(0,59)]


def sacar_5_bolitas():
    bolitas = []
    for i in range(5):
        bolitas.append(sacar_una_bolita())
    return bolitas


def contar_colores(bolitas):
    dic_colores = {}
    white = 0
    green = 0
    red = 0
    for i in range(len(bolitas)):
        if bolitas[i] == "White":
            white += 1
        elif bolitas[i] == "Green":
            green += 1
        else:
            red += 1
    return {"White":white,"Green":green,"Red":red}


contador1 = 0
contador2 = 0

for i in range(1000):
    dic = contar_colores(sacar_5_bolitas())
    if (dic["White"]== 3 and dic["Red"]== 2):
        contador1 +=1
    if (dic["White"]== 5 or dic["Red"]== 5 or dic["Green"]== 5 ):
        contador2 += 1


print('\nLa probabilidad experimental de tres blancas y dos rojas es '+ str(contador1/1000))
print('La probabilidad experimental de que las 5 bolitas sean del mismo color es '+str(contador2/1000))
```
